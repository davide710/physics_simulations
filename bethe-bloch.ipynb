```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
beta = np.concatenate((np.linspace(0.1, 0.8, 100), np.linspace(0.8, 0.999, 100), np.linspace(0.999, 0.99999, 100)))
T_p = 940 * (1/np.sqrt(1-beta**2) - 1) # T = mc^2 (gamma - 1)
T_mu = 100 * (1/np.sqrt(1-beta**2) - 1)
T_alpha = 940*4 * (1/np.sqrt(1-beta**2) - 1)
dEdx = 1/beta**2 * np.log(10**5 * beta**3/(1 - beta**2)) # 10^5 is completely arbitrary just to shift up the plot. dE/dx will be in arbitrary units but T will be correctly in MeV, and the minimum will be in the correct position
dEdx_alpha = 2**2 * 1/beta**2 * np.log(10**5 * beta**3/(1 - beta**2))
```

```python
plt.loglog(T_p, dEdx, label='proton')
plt.loglog(T_mu, dEdx, label='muon')
plt.loglog(T_alpha, dEdx_alpha, 'green', label='alpha particle')
plt.vlines(940 * (1/np.sqrt(1-0.96**2) - 1), 1, 100, 'blue', 'dashed', label='beta_p=0.96')
plt.vlines(100 * (1/np.sqrt(1-0.96**2) - 1), 1, 100, 'orange', 'dashed',label='beta_mu=0.96')
plt.vlines(940*4 * (1/np.sqrt(1-0.96**2) - 1), 1, 100, 'green', 'dashed',label='beta_alpha=0.96')
plt.legend()
plt.xlabel('T [MeV]')
plt.ylabel('dE/dx [arbitrary units]')
plt.grid()
plt.show()
```

```python
plt.plot(beta, dEdx)
plt.grid()
plt.xlabel('beta')
plt.ylabel('dE/dx [arbitrary units]')
plt.show()
```
